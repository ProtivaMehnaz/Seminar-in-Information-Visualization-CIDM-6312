'Collision_ID' column is not useful in analysis so droping the column from the dataset

```python
dfdrp = df.drop('COLLISION_ID', axis=1)
```

       'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST INJURED','NUMBER OF CYCLIST KILLED', 'NUMBER OF MOTORIST INJURED',
       'NUMBER OF MOTORIST KILLED' to get non-zero value.

```python
zeromask = (df['NUMBER OF PERSONS INJURED'] != 0) | (df['NUMBER OF PERSONS KILLED'] != 0) | (df['NUMBER OF PEDESTRIANS INJURED'] != 0) | (df['NUMBER OF PEDESTRIANS KILLED'] != 0) | (df['NUMBER OF CYCLIST INJURED'] != 0) | (df['NUMBER OF CYCLIST KILLED'] != 0) | (df['NUMBER OF MOTORIST INJURED'] != 0) | (df['NUMBER OF MOTORIST KILLED'] != 0)
cln =df[zeromask]
cln.shape
```

<strong><font face="Arial" size="4" color="blue">Motor vehicle collisions by hour in New York City:</font></strong>


```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create a new column for the month
cleandf['TIME'] = pd.to_datetime(cleandf['CRASH TIME'])
cleandf['HOUR'] = cleandf['TIME'].dt.hour


# Group the data by hour
grouped_data = cleandf.groupby('HOUR').size().reset_index(name='count')

# Create a line plot using the Seaborn library
sns.set(style='whitegrid', font_scale=1.2)
fig, ax = plt.subplots(figsize=(12, 8))
sns.barplot(data=grouped_data, x='HOUR', y='count', color='green', ax=ax)
ax.set_xlabel('HOUR', fontsize=14)
ax.set_ylabel('Number of Accidents', fontsize=14)
ax.set_title('Motor Vehicle Collisions/Crashes in New York City by HOUR', fontsize=18)
plt.show()
```

<strong><u>Analysis:</u></strong> We can see that around 4 in the afternoon the collision rate gets higher. This is the time of the day when usually people come back home from their workplaces, some people have a rush to pick up their children from the school/daycare on their way home. Along with that, around 8 in the morning, the number of accident is also higher. We can see that, this is the time usually people rush to go to their workplaces, or some people are going home after a night shift job, or having a  whole night party.

<strong><font face="Arial" size="4" color="blue">Motor vehicle collisions by month in New York City:</font></strong>

```python
# Create a new column for the month
cleandf['DATE'] = pd.to_datetime(cleandf['CRASH DATE'])
cleandf['YEAR'] = cleandf['DATE'].dt.year
cleandf['MONTH'] = cleandf['DATE'].dt.month

# Group the data by month
grouped_data = cleandf.groupby('MONTH').size().reset_index(name='count')

# Create a line plot using the Seaborn library
sns.set(style='whitegrid', font_scale=1.2)
fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(data=grouped_data, x='MONTH', y='count', color='blue', ax=ax)
ax.set_xlabel('Month', fontsize=14)
ax.set_ylabel('Number of Accidents', fontsize=14)
ax.set_title('Motor Vehicle Collisions/Crashes in New York City by Month', fontsize=18)
plt.show()
```

<strong><u>Analysis:</u></strong> We can observe from this visualization that in the month of October in each year the number of car crash is at the peak. Usually in the month of February and April the number goes down. 

<strong><font face="Arial" size="4" color="blue">Motor vehicle collisions by year in New York City:</font></strong>

```python
grouped_data = cleandf.groupby('YEAR').size().reset_index(name='count')
sns.set(style='whitegrid', font_scale=1.2)
fig, ax = plt.subplots(figsize=(12, 8))
sns.lineplot(data=grouped_data, x='YEAR', y='count', color='RED', ax=ax)
ax.set_xlabel('Years', fontsize=14)
ax.set_ylabel('Number of Accidents', fontsize=14)
ax.set_title('Motor Vehicle Collisions/Crashes in New York City by Year', fontsize=18)
plt.show()
```

<strong><u>Analysis:</u></strong> We can observe from this line plot that the year 2016 and 2018 were the year of highest number of motor vehicle accidents. Over 2018, there is a significant decline of accidents. Around the year 2022 the number was almost same as 2012, and then there is a steep decline. We can also observe that, after 2019, this was a period of a Covid-19 pandemic. Lockdown and other factors can influence the decrease of the car accidents.

To make visuals more explanatory, added two columns in to the dataset. "Total Killed" column will have sum of 'Number of Persons killed', 'Number of Pedestrians killed', 'Number of cyclist killed', 'number of Motorist killed' columns, and "Total injured" column will have sum of  'Number of Persons injured', 'Number of Pedestrians injured', 'Number of cyclist injured', 'number of Motorist injured'


```python
cleandf['Total Killed'] = cleandf.loc[:,['NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST KILLED', 
       'NUMBER OF MOTORIST KILLED']].sum(axis=1)
cleandf['Total Injured'] = cleandf.loc[:,['NUMBER OF PERSONS INJURED','NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF CYCLIST INJURED',
      'NUMBER OF MOTORIST INJURED']].sum(axis=1)
```

Pie chars are showing total number of persons killed/injured in each Borough

```python
dfgrp = cleandf.groupby(by=cleandf["BOROUGH"]).sum()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(12,12)) #ax1,ax2 refer to your two pies

normKValues = (1-0.1) *(dfgrp["Total Killed"]-dfgrp['Total Killed'].min())/(dfgrp['Total Killed'].max()-dfgrp['Total Killed'].min()) + 0.1
noriKValues = (1-0.1) *(dfgrp["Total Injured"]-dfgrp['Total Injured'].min())/(dfgrp['Total Injured'].max()-dfgrp['Total Injured'].min()) + 0.1

cmap = plt.get_cmap("Blues")
colorsK = cmap(normKValues)

cmap = plt.get_cmap("Greens")
colorsI = cmap(noriKValues)

# create white circles at the center of the pie charts
circle1 = patches.Circle((0,0), 0.5, color='white')
circle2 = patches.Circle((0,0), 0.5, color='white')

ax1.add_artist(circle1) # add circle to ax1
ax1.pie(dfgrp["Total Killed"],labels=dfgrp.index, colors = colorsK, autopct = '%1.1f%%', wedgeprops=dict(width=0.3, edgecolor='w')) #plot first pie
ax1.set_title("Total Killed by Borough")

ax2.add_artist(circle2) # add circle to ax2
ax2.pie(dfgrp["Total Injured"],labels=dfgrp.index,colors = colorsI, autopct = '%1.1f%%', wedgeprops=dict(width=0.3, edgecolor='w')) #plot second pie
ax2.set_title('Total Injured by Borough')

fig.tight_layout(pad=2.0) 
plt.subplots_adjust(wspace =0.7, hspace=0.3)

plt.show()
```

<strong><font face="Arial" size="4" color="blue">Motor vehicle collision Heatmap in Manhattan city:</font></strong>

```python
import warnings
locations = cleandf.groupby(['LATITUDE', 'LONGITUDE']).size().reset_index(name='count')
# Define the latitude and longitude range for Manhattan
lat_range = (40.7000, 40.8800)
long_range = (-74.0200, -73.9100)

# Select rows where the latitude and longitude are within the specified range
subset = cleandf.loc[(cleandf['LATITUDE'] >= lat_range[0]) & (cleandf['LATITUDE'] <= lat_range[1]) & 
                     (cleandf['LONGITUDE'] >= long_range[0]) & (cleandf['LONGITUDE'] <= long_range[1])]

# Aggregate the number of accidents by location
locations = subset.groupby(['LATITUDE', 'LONGITUDE']).size().reset_index(name='count')

# Create a heatmap using the Seaborn library
sns.set(style='white', font_scale=1.2)
fig, ax = plt.subplots(figsize=(12,8))
sns.kdeplot(data=locations, x='LONGITUDE', y='LATITUDE', cmap='Reds', shade=True, thresh=0.05, alpha=0.5, 
            kernel='epa', bw_adjust=0.3, n_levels=200, ax=ax)
ax.set_xlabel('Longitude', fontsize=14)
ax.set_ylabel('Latitude', fontsize=14)
ax.set_title('Motor Vehicle Collisions/Crashes in Manhattan', fontsize=18)
plt.show()
```
