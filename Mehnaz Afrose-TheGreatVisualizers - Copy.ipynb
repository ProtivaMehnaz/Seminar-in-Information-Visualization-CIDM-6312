To make visuals more explanatory, added two columns in to the dataset. "Total Killed" column will have sum of 'Number of Persons killed', 'Number of Pedestrians killed', 'Number of cyclist killed', 'number of Motorist killed' columns, and "Total injured" column will have sum of  'Number of Persons injured', 'Number of Pedestrians injured', 'Number of cyclist injured', 'number of Motorist injured'


```python
cleandf['Total Killed'] = cleandf.loc[:,['NUMBER OF PERSONS KILLED', 'NUMBER OF PEDESTRIANS KILLED', 'NUMBER OF CYCLIST KILLED', 
       'NUMBER OF MOTORIST KILLED']].sum(axis=1)
cleandf['Total Injured'] = cleandf.loc[:,['NUMBER OF PERSONS INJURED','NUMBER OF PEDESTRIANS INJURED', 'NUMBER OF CYCLIST INJURED',
      'NUMBER OF MOTORIST INJURED']].sum(axis=1)
```

Pie chars are showing total number of persons killed/injured in each Borough

```python
dfgrp = cleandf.groupby(by=cleandf["BOROUGH"]).sum()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

fig, (ax1,ax2) = plt.subplots(1,2,figsize=(12,12)) #ax1,ax2 refer to your two pies

normKValues = (1-0.1) *(dfgrp["Total Killed"]-dfgrp['Total Killed'].min())/(dfgrp['Total Killed'].max()-dfgrp['Total Killed'].min()) + 0.1
noriKValues = (1-0.1) *(dfgrp["Total Injured"]-dfgrp['Total Injured'].min())/(dfgrp['Total Injured'].max()-dfgrp['Total Injured'].min()) + 0.1

cmap = plt.get_cmap("Blues")
colorsK = cmap(normKValues)

cmap = plt.get_cmap("Greens")
colorsI = cmap(noriKValues)

# create white circles at the center of the pie charts
circle1 = patches.Circle((0,0), 0.5, color='white')
circle2 = patches.Circle((0,0), 0.5, color='white')

ax1.add_artist(circle1) # add circle to ax1
ax1.pie(dfgrp["Total Killed"],labels=dfgrp.index, colors = colorsK, autopct = '%1.1f%%', wedgeprops=dict(width=0.3, edgecolor='w')) #plot first pie
ax1.set_title("Total Killed by Borough")

ax2.add_artist(circle2) # add circle to ax2
ax2.pie(dfgrp["Total Injured"],labels=dfgrp.index,colors = colorsI, autopct = '%1.1f%%', wedgeprops=dict(width=0.3, edgecolor='w')) #plot second pie
ax2.set_title('Total Injured by Borough')

fig.tight_layout(pad=2.0) 
plt.subplots_adjust(wspace =0.7, hspace=0.3)

plt.show()
```
